## Report: act_report

WeRateDogs is an exciting twitter page that rates dogs of different breeds and stages professionally. A feature unique to these ratings is how dogs are rated above the rate-scale which is 10. These ratings can be as high as 15, even more. 

An archive of WeRateDogs' tweets which was readily available, a dataset which contains dog_breed prdictions by a neural network of the rated dogs and a last dataset that contains favorite and retweet count queried via Twitter API were wrangled to obtain a single and high-quality dataset which included original ratings only. An exploratory analysis was carried out on the final dataset to get some insights too. This led to asking the following questions;

1. **What year did the handler make the most ratings?**

Year 2016 had the highest number of ratingss of the three years. About 1000 dogs were rated in this year.

2. **What percentage of the dogs had a rating of more than 10?**

57.83% of the dogs rated had a rating of >10.

3. **Which are the top 5 most rated dogs breeds?**

The top 5 most rated breeds of dog are;
1.         golden_retriever     - 136
2.         Labrador_retriever   -  91
3.        Pembroke              - 88
4.        Chihuahua             - 78
5.        pug                   - 54

4. **Which dog stage was mostly rated?**

Most of the dogs rated were in the Pupper dog stage.
